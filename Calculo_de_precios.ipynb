        <h3>1. Introducción</h3>
            <h3>&emsp;<font color = 'black'>Una opción financiera es un contrato derivado que otorga al titular el derecho, pero no la obligación, de comprar o vender un activo subyacente a un precio preestablecido (llamado precio de ejercicio o strike price) en o antes de una fecha futura específica. El activo subyacente puede ser una acción, una divisa, un índice o un bien tangible, y es el objeto que determina el valor de la opción. Existen dos tipos principales de opciones: las opciones de compra (call) y las opciones de venta (put).Una opción de compra (call) otorga al titular el derecho de comprar el activo subyacente a un precio fijo, mientras que una opción de venta (put) concede el derecho de vender el activo al precio de ejercicio. El precio que paga el comprador de la opción por obtener estos derechos se conoce como prima de la opción.
            <h3>Una distinción importante entre opciones es su estilo de ejercicio. Las opciones europeas solo se pueden ejercer en la fecha de vencimiento, mientras que las opciones americanas pueden ejercerse en cualquier momento antes de la fecha de vencimiento. Esta diferencia es crucial en la toma de decisiones de los inversionistas, ya que las opciones europeas limitan la flexibilidad en comparación con las americanas.
            <h3>Si el titular de una opción europea (ya sea tipo call o put) decide no ejercer su derecho en la fecha de vencimiento, la opción simplemente expira sin valor, y no se incurre en mayores pérdidas más allá de la prima pagada por la opción.
            <h3> Por ejemplo, si un inversionista posee una opción de compra europea sobre acciones de Tesla con un precio de ejercicio de $50 y la fecha de vencimiento es el 1 de enero de 2024, su decisión dependerá del precio de las acciones en esa fecha. Si el precio de las acciones de Tesla es inferior a $50 en la fecha de vencimiento, el inversionista no ejercería su opción, ya que sería más económico comprar las acciones directamente en el mercado. Sin embargo, si el precio de las acciones es superior a $50, el inversionista ejercería su opción para adquirir las acciones a un precio más bajo que el del mercado.
            <h3>En otro caso, supongamos que un inversionista tiene una opción de venta europea sobre acciones de Nvidia con un precio de ejercicio de $75 y la fecha de vencimiento es el 1 de julio de 2024. Si el precio de las acciones de Nvidia es superior a $75 en la fecha de vencimiento, el inversionista no ejercería la opción, ya que sería más rentable vender las acciones en el mercado abierto. Si el precio de las acciones es inferior a $75, el inversionista ejercería su opción para vender las acciones a un precio más alto del que podría obtener en el mercado.
            <h3>Otro aspecto relevante es el impacto de los dividendos. Un dividendo es un pago que una empresa realiza a sus accionistas y puede afectar el valor de una opción. Las empresas que no pagan dividendos tienden a tener un menor impacto en la prima de las opciones, ya que el pago de dividendos tiende a reducir el precio de las acciones.
            <h3>La volatilidad es un concepto central en la valoración de opciones y se refiere a la magnitud de las fluctuaciones en el precio del activo subyacente. En el modelo de Black-Scholes, se asume que la volatilidad es constante y que los precios de los activos siguen un movimiento browniano geométrico. Este modelo ha sido fundamental en las finanzas cuantitativas, aunque presenta algunas limitaciones, como la suposición de volatilidad constante y la falta de capacidad para capturar eventos extremos o shocks de mercado.
            <h3> Las variables clave en el modelo de Black-Scholes para la valoración de opciones europeas incluyen el precio del activo subyacente, el precio de ejercicio, el tiempo hasta la maduración, la volatilidad del activo, la tasa de interés libre de riesgo y la volatilidad implícita. La volatilidad implícita es la expectativa del mercado sobre la futura volatilidad del activo subyacente, derivada del precio de la opción en el mercado.
            <h3> El factor exponencial e^rT en la fórmula de Black-Scholes representa el valor presente del precio de ejercicio, descontado a la tasa de interés libre de riesgo, r, durante el tiempo restante hasta la maduración, T. Un aumento en la tasa de interés libre de riesgo incrementa el precio de las opciones de compra (call) y reduce el de las opciones de venta (put), ya que eleva el valor presente del precio de ejercicio.
            <h3> Por otro lado, un aumento en la volatilidad del activo subyacente generalmente aumenta el precio de ambas opciones (call y put), ya que incrementa la probabilidad de que la opción tenga valor en el futuro. Si el precio de ejercicio aumenta, el valor de una opción de compra disminuye, mientras que el de una opción de venta aumenta. Finalmente, a medida que disminuye el tiempo hasta la maduración (T), el valor de la opción también disminuye, un fenómeno conocido como decadencia temporal.

        <h3><i>2.1 Modelos de Valoración de Opciones</i></h3>
        <h3> Uno de los modelos más reconocidos para la valoración de opciones europeas es la ecuación de Black-Scholes. Este modelo fue desarrollado por Fischer Black y Myron Scholes en 1973 y revolucionó el análisis financiero al proporcionar una fórmula para calcular el precio de opciones sin tener que recurrir a la simulación directa del activo subyacente. La ecuación de Black-Scholes se basa en varios supuestos, tales como la suposición de que los precios de los activos siguen un Movimiento Browniano Geométrico, lo cual implica que la volatilidad es constante y que el mercado es eficiente.
<h3> El modelo de Black-Scholes se ha aplicado extensamente debido a su capacidad para simplificar el proceso de valoración, permitiendo que los precios de las opciones dependan únicamente de factores observables como el precio del activo subyacente, el precio de ejercicio, la volatilidad, el tiempo hasta la expiración y la tasa de interés libre de riesgo. No obstante, a pesar de su relevancia, una limitación importante es la incapacidad del modelo para capturar fenómenos de volatilidad variable, lo que ha impulsado el desarrollo de métodos alternativos como los modelos ARCH y GARCH, los cuales permiten modelar la volatilidad de manera condicional. </3>
        

<font face="arial" size="3" color="black">
    <div align="justify">
        <h3><i>2.2  Métodos Estocásticos y Simulación de Monte Carlo</i></h3>
        <h3>&emsp;Además de los modelos deterministas como Black-Scholes, los métodos estocásticos han demostrado ser herramientas poderosas para modelar la incertidumbre y volatilidad en los mercados financieros. En este contexto, las cadenas de Markov y la simulación de Monte Carlo se utilizan frecuentemente para realizar predicciones probabilísticas del comportamiento de los precios de los activos subyacentes. Una cadena de Markov es un proceso estocástico que asume que la probabilidad de transición de un estado al siguiente depende únicamente del estado actual y no de los eventos pasados, lo cual lo convierte en un modelo útil para la simulación de trayectorias de precios de activos.

<h3> Por otro lado, la simulación de Monte Carlo es un método numérico que genera múltiples escenarios posibles para un activo financiero mediante la generación de caminos aleatorios que simulan la evolución del precio del activo subyacente. En el caso de la valoración de opciones, esta técnica es fundamental para evaluar la incertidumbre inherente en los mercados financieros, ya que permite estimar el valor esperado del precio de una opción bajo distintas trayectorias posibles del activo. En este sentido, la combinación de modelos ARCH para capturar la volatilidad condicional junto con la simulación de Monte Carlo ofrece una aproximación robusta para la valoración de opciones en mercados con fluctuaciones impredecibles.

### Movimiento Browniano Geométrico

Un proceso estocástico $(S_t)_{t\geq 0}$ sigue un movimiento browniano geométrico si satisface la ecuación diferencial estocástica:

$$dS_t = \mu S_t dt + \sigma S_t dW_t$$

donde el proceso $(W_t)_{t\geq 0}$ sigue un movimiento browniano estándar. La solución de esta ecuación está dada por

$$\displaystyle S_t=S_0\,e^{(\mu - \sigma^2/2)\,t+\sigma\,W_t}$$

Para un activo financiero, $S_t$ es el precio del activo en el tiempo $t$, $S_0$ es el precio inicial del activo en $t=0$, $\mu$ es la tasa de crecimiento esperada y $\sigma$ es la volatilidad.

````python
def gbm_2(nsteps=1000, t=1, mu=0.001, sigma=0.02, start=1):
    dt = t / nsteps
    steps = [(mu - 0.5 * sigma**2) * dt + np.random.randn() * sigma * np.sqrt(dt) for i in range(nsteps)]
    y = start * np.exp(np.cumsum(steps))
    x = [dt * i for i in range(nsteps)]
    return x, y
````

<font face="times new roman" size="3" color="black">
    <h3> La fórmula de Black-Scholes la utilizamos para determinar cual va a ser el precio de la opción call. Utilizamos las siguientes entradas.
S0: Precio actual del activo
K: Precio de ejercicio
T: Tiempo hasta el vencimiento(1 año en nuestro caso)
r: Tasa de interes libre de riesgo
sigma: Volatilidad del activo

````python
# Fórmula de Black-Scholes
def black_scholes_call(S0, K, T, r, sigma):
    d1 = (np.log(S0 / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S0 * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return call_price

def black_scholes_put(S, K, T, r, sigma):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    put_price = K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
    return put_price
````

````python
call_price = black_scholes_call(S0, K, T, r, sigma)
print(f"Valor de la opción Call: {call_price}")

put_price = black_scholes_put(S0, K, T, r, sigma)
print(f"Valor de la opción Put: {put_price}")
````

###Modelo ARCH

En esta sección ajustamos un moedlo ARCH(Autoregressive Conditional Heteroskedasticity) a los rendimientos logaritmicos para asi poder modelar la volatilidad

````python
model_arch = arch_model(np.array(log_returns), vol='ARCH', p=1)
arch_fit = model_arch.fit(disp='off')
````

````python
# Obtener los parámetros ajustados
params = arch_fit.params
omega = params['omega']
alpha = params['alpha[1]']

# Simular rendimientos futuros
n_simulations = 10000  # Número de simulaciones
n_periods = 252  # Número de días a simular (1 año)

# Inicializar matrices para guardar los rendimientos y las volatilidades simuladas
simulated_returns = np.zeros((n_simulations, n_periods))
simulated_volatility = np.zeros((n_simulations, n_periods))
````

````python
np.random.seed(42)
for i in range(n_simulations):
    # Simular el primer valor
    epsilon = np.random.normal(0, np.sqrt(omega))
    simulated_returns[i, 0] = epsilon
    simulated_volatility[i, 0] = np.sqrt(omega)

    # Simular el resto de los valores
    for t in range(1, n_periods):
        volatility_t = np.sqrt(omega + alpha * simulated_returns[i, t-1]**2)
        simulated_returns[i, t] = volatility_t * np.random.normal(0, 1)
        simulated_volatility[i, t] = volatility_t
````

### Cadenas de Markov

````python
n_simulations = 10000

# Generate random price paths using a Markov chain-like process
def simulate_stock_price(S0, T, r, sigma, n_steps):
    dt = T / n_steps
    prices = np.zeros(n_steps)
    prices[0] = S0
    for t in range(1, n_steps):
        z = np.random.standard_normal()
        prices[t] = prices[t-1] * np.exp((r - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * z)
    return prices

# Monte Carlo simulation to estimate call and put option prices
def monte_carlo_option(S0, K, T, r, sigma, n_simulations, n_steps):
    call_payoffs = []
    put_payoffs = []

    for _ in range(n_simulations):
        prices = simulate_stock_price(S0, T, r, sigma, n_steps)
        # Call option payoff
        call_payoff = max(prices[-1] - K, 0)
        call_payoffs.append(call_payoff)

        # Put option payoff
        put_payoff = max(K - prices[-1], 0)
        put_payoffs.append(put_payoff)

    # Discount back to present value
    call_option_price = np.exp(-r * T) * np.mean(call_payoffs)
    put_option_price = np.exp(-r * T) * np.mean(put_payoffs)

    return call_option_price, put_option_price

call_option_price, put_option_price = monte_carlo_option(S0, K, T, r, sigma, n_simulations, 252)

# Print results
print(f"Call Option Price: {call_option_price:.2f}")
print(f"Put Option Price: {put_option_price:.2f}")
````
